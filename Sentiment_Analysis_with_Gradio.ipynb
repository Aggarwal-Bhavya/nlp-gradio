```python
import gradio as gr #importing gradio module
import re #importing RegEx module

from transformers import pipeline #from Hugging Face's transformers, we import pipeline module
```

Defining sentiment analysis function

---

```python
sentiment = pipeline("sentiment-analysis")

def sentiment_analysis(text): 
  text = text.encode("ascii", errors = "ignore").decode(
      "ascii"
  ) #remove non-ascii characters
  text = text.lower() #converts tthe text to lower case
  text = re.sub(r"\n", " ", text) #removing next line
  text = re.sub(r"\n\n", " ", text) #in re.sub(), r is part of string syntax, if not used, \n has to be written as \\n
  text = re.sub(r"\t", " ", text) #removes tab or continuous spaces
  text = text.strip(" ") #trims the text of extra spaces at beginning and end of text
  text = re.sub(r"[^\w\s]", "", text) #remove punctuation and special characters
  #[^\w\s] this regex is used to denote any character that does not belong to either the word or whitespace group
  text = re.sub(" +", " ", text).strip()
  #gets rid of multiple spaces and replaces with a single space
  results = sentiment(text) #passes the processed text in pipeline of sentiment analysis
  return results[0]["label"], round(results[1]["score"], 5) #the result consists of two components label and score
```

Defining gradio interface

---

```python
app = gr.Interface(
    fn = sentiment_analysis,
    title = "Sentiment Analysis",
    description = "Enter some text and see if the distributed model can predict the sentiment accurately",
    inputs = gr.inputs.Textbox(lines = 10, label = "Enter some text here"),
    outputs = [
               gr.outputs.Textbox(label = "Sentiment Label"),
               gr.outputs.Textbox(label = "Sentiment Score")
    ]
)
```

```python
app.launch(share=True)
```
